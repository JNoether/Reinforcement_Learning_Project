```python
import json
import os
import matplotlib.pyplot as plt
```

```python
with open("datasets/data/train/seq/0_seq.json") as file:
    grid = json.load(file)
    print(grid)
    
with open("datasets/data/train/task/0_task.json") as file:
    grid = json.load(file)
    print(grid)
```

```python
# plots for grids
walls = []
pre_markers = []
post_markers = []
diff_markers = []

# compute values
for seq in os.listdir("datasets/data/train/task"):
    with open("datasets/data/train/task/" + seq) as file:
        grid = json.load(file)
        walls.append(len(grid["walls"]))
        pre_markers.append(len(grid["pregrid_markers"]))
        post_markers.append(len(grid["postgrid_markers"]))
        diff_markers.append(len(grid["pregrid_markers"]) - len(grid["postgrid_markers"]))
        
        
# generate plots
fig = plt.figure()
plt.hist(walls)
plt.title("amount of walls of grid")

fig= plt.figure()
plt.hist(pre_markers)
plt.title("amount of markers on the pre-grid")

fig= plt.figure()
plt.hist(post_markers)
plt.title("amount of markers on the post-grid")

fig= plt.figure()
plt.hist(diff_markers)
plt.title("difference between the amount of markers in the pre- and post-grid")
```

```python
# plots of the sequences
lengths = []
for seq in os.listdir("datasets/data/train/seq"):
    with open("datasets/data/train/seq/" + seq) as file:
        grid = json.load(file)
        lengths.append(len(grid["sequence"]))

        
plt.hist(lengths)
plt.title("length of an optimal episode")
```
